```python
import numpy as np
from scipy.optimize import curve_fit as cf
import matplotlib.pyplot as plt
```

```python
# Defining a linear model

def linear(x, m, c):
    return x*m + c
```

```python
# Creating some random data
xdata = np.arange(0, 4, 0.01)
random = np.random.randn(len(xdata))
ydata = linear(xdata, 1.2, 1.5) + random

# Plotting the data
plt.scatter(xdata, ydata, s=1)
```

```python
# Applying curve fit to get the optimized values and the covariance matrix

p_opt, p_cov = cf(linear, xdata, ydata)
print(p_opt, p_cov)
```

```python
# Getting the new ydata
new_ydata = linear(xdata, *p_opt)
```

```python
# Plotting the new ydata

plt.plot(xdata, new_ydata, label = "Best Fit", color='red')
plt.scatter(xdata, ydata, s=1)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
# Considering a quadratic equation

def quadratic(x, a, b, c):
    return a*x**2 + b*x + c
```

```python
# Creating the data

quad_y = quadratic(xdata, 1, 4, 4) + np.random.randn(len(xdata))
```

```python
# Fitting the curve

p_opt_quad, p_cov_quad = cf(quadratic, xdata, quad_y)
new_quad_y = quadratic(xdata, *p_opt_quad)

# Plotting the curve and the original

plt.plot(xdata, new_quad_y, color='red', label="Best Fit")
plt.scatter(xdata, quad_y, s=1)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
# Finding the error in the parameters

print(f"Error in a = {abs(p_opt_quad[0] - 1)}, b = {abs(p_opt_quad[1] - 4)}, c = {abs(p_opt_quad[2] - 4)}")
```
