```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
url="https://raw.githubusercontent.com/krittikaiitb/tutorials/2878a43df2b210fa5cb15aff2b1efa68d5478439/Tutorial_4/GW_data_file.csv"
strain,freq,time=np.loadtxt(url,delimiter=',')

#strain vs time plot
plt.plot(time,strain)
plt.xlabel("Time")
plt.ylabel("Strain Data")
plt.title("Strain Data vs Time")
plt.grid()
plt.show()

#freq vs time plot
plt.plot(time,freq)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.title("Frequency vs Time")
plt.grid()
plt.show()

#freq vs time at zero crossings
zeroindices=[]
for i in range(len(strain)-1):
  if (strain[i]==0 and strain[i+1]!=0) or (strain[i]==0 and strain[i+1]==0):
    zeroindices.append(i)
  elif (strain[i]>0 and strain[i+1]<0) or (strain[i]<0 and strain[i+1]>0):
    zeroindices.append(i)
  #elif strain[i]==0 and strain[i+1]==0:
    #zeroindices.append(i)
#print(zeroindices)
plt.scatter(time[zeroindices],freq[zeroindices],s=10,color='indigo')
plt.xlabel("Zero Crossings Time", color='white')
plt.ylabel("Zero Crossings Frequency")
plt.title("Zero Crossings Frequency vs Time")
plt.grid()
#plt.show()
```
